After cleansing data, now we want to plot graph.
Lets start with an example to understand the structure needed to plot graph.

```ruby
sample_data = {"train": [1, 2],
       "train2": [2, 3.4]}
sample_df = Daru::DataFrame.new(sample_data, index: Daru::Index.new([:city, :city2], name: 'idx_city'))
```

```ruby
sample_df[:train]
```

```ruby
# Need to use daru view, so that chart can show up in jupyter notebook
Daru::View.plotting_library = :nyaplot

sample_chart_data = sample_df[:train]
sample_chart_data.plot(type: :bar) do |plot|
  plot.width 1200
end
```

Now we organize the data, so that we can plot the graph for time taken for train ES9051

```ruby
# if you want to replace nil with something else like 0, you can use replace_nils
chart_data = Daru::Vector.new(df[:es9051], index: df[0].to_a).reject_values nil, Float::NAN

require 'time'
time_data = []
chart_data.each_entry do |e|
  time_data << (Time.parse(e) - Time.parse(chart_data[0]))/60 /60
end
chart_data = Daru::Vector.new(time_data, index: chart_data.index)
# type (:scatter, :bar, :histogram)
chart_data
```

```ruby
chart_data.plot(type: :bar).configure do
  #width(1080)
  height(1200)
end
```

```ruby
chart_data.plot(type: :bar).to_iruby
```

```ruby
plt = Daru::View::Plot.new(chart_data, type: :bar)
plt.show_in_iruby
```

Lets do some 3D plotting!

```ruby
require 'nyaplot3d'
plot = Nyaplot::Plot3D.new
['#ff7f00','#1f78b4','#a6cee3'].each_with_index do |color, index|
  x=[];y=[];z=[];dz = 5*rand
  0.step(1, 0.2) do |i|
    0.step(1, 0.2) do |j|
      x.push(i+rand)
      y.push(j+rand)
      z.push(Math.sin(i)*Math.sin(j)+dz+rand)
    end
  end
  p = plot.add(:particles, x, y, z)
  p.color(color)
  p.name('molecules')
end
plot.export_html
```

Go to your notebook workspace, and you will see the graph generated html.
